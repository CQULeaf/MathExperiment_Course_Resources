## 问题二：回忆线性代数基本概念与运算并用代码实现

1. **方阵的行列式 (Determinant)**
   - **概念**：方阵的行列式是一个**标量值**，它反映了方阵的某些性质。行列式的值可以用来判断方阵是否可逆，**若行列式为0，则方阵不可逆**。
   
2. **矩阵的秩 (Rank)**
   - **概念**：矩阵的秩是指**矩阵行向量或列向量的最大线性无关组中包含向量的个数**。秩表示了矩阵中包含的信息量，也决定了矩阵解线性方程组的性质。

3. **方阵的幂 (Matrix Power)**
   - **概念**：方阵的幂是指将**方阵与自身相乘多次**。例如，一个方阵A的二次幂就是A乘以A。

4. **方阵的逆阵 (Inverse Matrix)**
   - **概念**：如果两个方阵A和B满足**AB=BA=I（I是单位矩阵）**，则我们称B是A的逆阵。只有当方阵的行列式不为0时，逆阵才存在。

5. **向量组的秩**
   - **概念**：与矩阵的秩类似，向量组的秩是指**该向量组中最大的线性无关子集的大小**。

6. **向量组的正交化 (Vector Orthogonalization)**
   - **概念**：向量组的正交化是**将一组向量转换成正交向量组**的过程，通常使用格拉姆-施密特正交化过程。

7. **方阵的特征向量和特征值 (Eigenvalues and Eigenvectors)**
   - **概念**：对于一个方阵A，如果存在一个非零向量v和一个标量λ，使得**Av=λv**，那么v是A的一个特征向量，λ是对应的特征值。

```python
A = np.random.rand(4, 4) # 创建一个4x4的随机方阵

detA = np.linalg.det(A) # 方阵的行列式
rankA = np.linalg.matrix_rank(A) # 矩阵的秩
A_squared = np.linalg.matrix_power(A, 3) # 方阵的幂
invA = np.linalg.inv(A) # 方阵的逆秩

# 向量组的秩
# 此处直接使用矩阵的秩，因为向量组可以视为矩阵的行或列
rank_vectors = np.linalg.matrix_rank(A)

# 向量组的正交化
# 以方阵A的列向量为例进行Gram-Schmidt正交化
# Q为正交矩阵且矩阵的列向量是原矩阵列向量的正交化版本，R为上三角矩阵
Q, R = np.linalg.qr(A)

eigenvalues, eigenvectors = np.linalg.eig(A)

print(f"Determinant: {detA:.3f}")
print("Rank:", rankA)
print("Matrix Power (A^2):", A_squared)
print("Inverse Matrix:", invA)
print("Orthogonalization (First Vector):", Q[:,0])
print("Eigenvalues:", eigenvalues)
print("First Eigenvector:", eigenvectors[:,0])
```

## 问题三：验证行列式展开定理及克莱姆法则

1. 验证行列式的展开定理

    行列式的展开定理指出，矩阵 $ A $ 的行列式可以通过选择任意一行（或列），将该行（或列）的每个元素与其对应的代数余子式相乘，然后将这些乘积求和来计算。

2. 验证克莱姆法则

    克莱姆法则说明，如果一个线性系统的系数矩阵的行列式非零，那么该系统有唯一解，并且每个未知数可以通过将系数矩阵的对应列替换为等式右边的常数列，并计算这个新矩阵的行列式，然后除以原矩阵的行列式来求解。

```python
import numpy as np

# 验证行列式的展开定理
def verify_determinant_expansion(A):
    det_A = np.linalg.det(A) # 计算行列式
    expansion_sum = 0
    for i in range(len(A)):
        minor = np.linalg.det(np.delete(np.delete(A, 0, axis=0), i, axis=1))  # 删除第0行和第i列
        cofactor = ((-1)**(i)) * minor # 计算代数余子式
        expansion_sum += A[0, i] * cofactor # 展开定理求和
    close = np.isclose(det_A, expansion_sum) # 检查两个值是否相等
    return det_A, expansion_sum, close

# 验证克莱姆法则
def verify_cramers_rule(A, b):
    det_A = np.linalg.det(A) # 计算系数矩阵的行列式
    # 如果行列式为0，无解
    if det_A == 0:
        return "System does not have a unique solution"
    # 计算每个未知数的解
    solutions = []
    for i in range(len(A)):
        Ai = np.copy(A)
        Ai[:, i] = b
        det_Ai = np.linalg.det(Ai)
        x_i = det_Ai / det_A
        solutions.append(x_i)
    return solutions

# 示例矩阵与向量
A = np.array([[2, 1, -1],
              [-3, -1, 2],
              [-2, 1, 2]])

b = np.array([8, -11, -3])

# 验证行列式的展开定理
det_A, expansion_sum, close_1 = verify_determinant_expansion(A)

# 实现克莱姆法则
try:
    solutions = verify_cramers_rule(A, b)
    solutions_rounded = np.around(solutions, decimals=3)
except ValueError as e:
    solutions = str(e)

# 使用NumPy的线性方程组求解函数作为验证
numpy_solution = np.linalg.solve(A, b)

# 检查两种方法的解是否相等
close_2 = np.allclose(solutions, numpy_solution)

print(f"Determinant: {det_A:.3f}")
print(f"Expansion Sum: {expansion_sum:.3f}")
print(f"Determinant and Expansion Sum Are Close: {close_1}")
print(f"Cramer's Rule Solutions: {solutions_rounded}")
print(f"NumPy Solution: {numpy_solution}")
print(f"Cramer's Rule Solutions and NumPy Solution Are Close: {close_2}")
```

## 问题五：计算所有的3-6位自幂数

自幂数的定义：它们等于其各位数字的幂次之和，幂次等于它们的位数。

```python
def is_narcissistic_number(num):
    # 将数字转换为字符串以便迭代每个数字
    num_str = str(num)
    num_length = len(num_str)
    # 计算每个数位的幂之和
    sum_of_powers = sum(int(digit) ** num_length for digit in num_str)
    return sum_of_powers == num

# 计算所有3到6位的自幂数
narcissistic_numbers = {n: [] for n in range(3, 7)}
for i in range(3, 7):
    # 生成3到6位数的可能范围
    start = 10**(i - 1)
    end = 10**i
    # 检查每个数是否为自幂数
    for num in range(start, end):
        if is_narcissistic_number(num):
            narcissistic_numbers[i].append(num)

print("3-Digit Narcissistic Numbers:", narcissistic_numbers[3])
print("4-Digit Narcissistic Numbers:", narcissistic_numbers[4])
print("5-Digit Narcissistic Numbers:", narcissistic_numbers[5])
print("6-Digit Narcissistic Numbers:", narcissistic_numbers[6])
```

## 问题六：计算圆周率


编写一个程序来验证22/7（约率）和355/113（密率）是否是在两位数分数和三位数分数中分别最接近圆周率π的值，并要求找出四位数分数中最接近π的值。
同时回答如何优化程序以提高其效率，并要求简要解释程序的优点。

```python
from fractions import Fraction

# 函数用于找出最接近圆周率的分数
def find_closest_fraction(denom_limit):
    pi = np.pi
    closest_fraction = Fraction(0)
    min_difference = pi  # 初始化为pi，保证第一个差值肯定更小
    # 从1遍历到denom_limit作为分母
    for denominator in range(1, denom_limit):
        # 分子是圆周率乘以分母后取最接近的整数
        numerator = round(pi * denominator)
        fraction = Fraction(numerator, denominator)
        difference = abs(fraction - pi)
        if difference < min_difference:
            min_difference = difference
            closest_fraction = fraction
    return closest_fraction

# 验证22/7和355/113
pi_approximations = {
    2: find_closest_fraction(10**2),
    3: find_closest_fraction(10**3)
}

# 寻找四位数分数中最接近于圆周率的数
pi_approximations[4] = find_closest_fraction(10**4)

```

## 问题七：绘制函数的Taylor展开公式图像

使用函数的Taylor展开公式，得到函数 $ y = \exp(x)$ 在点 $x = 1$ 的前8次的多项式展开的图像，并得出一定的结论

函数 $ y = \exp(x)$ 在点 $x = 1$ 的泰勒展开式可以写为：

$$ \exp(x) = \exp(1) + \exp(1)(x - 1) + \frac{\exp(1)}{2!}(x - 1)^2 + \cdots + \frac{\exp(1)}{n!}(x - 1)^n + \cdots $$

为了得到前8次的多项式展开，我们只需计算到第7阶导数的项（因为我们从0阶项开始计算）。

```python
import math

# 定义原函数
def exp_x(x):
    return np.exp(x)

# 泰勒展开的多项式函数
def taylor_expansion(x, n):
    sum = 0
    for i in range(n + 1):
        sum += np.exp(1) * ((x - 1)**i) / math.factorial(i)
    return sum

mpl.rcParams['font.family'] = 'Palatino Linotype'

# 绘制图像
x = np.linspace(0, 2, 400)
y = exp_x(x)

plt.figure(figsize=(10, 6))

# 绘制原函数
plt.plot(x, y, label='exp(x)', linewidth=2)

# 绘制前8次的多项式展开
for n in range(8):
    y_taylor = taylor_expansion(x, n)
    plt.plot(x, y_taylor, label=f'Taylor Expansion (n={n})')

plt.title('Function exp(x) and its Taylor Expansions at x=1', fontsize=16)
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)
plt.legend(fontsize=13)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True)
plt.show()
```

通过绘制函数 $y = \exp(x)$ 及其在点 $x = 1$ 的前8次泰勒多项式展开的图像，可以得出以下结论：

1. **逼近效果逐渐提升**：随着展开次数的增加，泰勒多项式逐渐接近原函数 $y = \exp(x)$。这说明，泰勒展开是一个非常强大的工具，很适合用于逼近复杂函数。

2. **局部逼近**：泰勒展开是关于**某一点**的局部逼近。在这个例子中，我们选择 $x=1$ 作为展开点。可以看出，多项式在 $x=1$ 附近的逼近效果最好，而随着 $x$ 值偏离1，逼近效果逐渐变差。

3. **高阶展开的重要性**：初阶泰勒多项式（如一阶、二阶）对于函数的逼近可能不够精确，特别是在远离展开点的区域。随着展开次数的增加，多项式能够更准确地逼近原函数，这对于需要**高精度逼近**的应用场景非常重要。

## 问题八：验证Collatz猜想并统计步数

科拉兹猜想（Collatz 猜想），也被称为$3n + 1$猜想、奇偶归一猜想等，是指对于每一个正整数，如果它是偶数，则将它除以2；如果它是奇数，则将它乘以3并加上1。重复进行这样的操作，最终总能够得到1。

```python
def collatz_steps(n):
    steps = 0
    while n != 1:
        n = n / 2 if n % 2 == 0 else 3 * n + 1
        steps += 1
    return steps

## 给出一个数8888，计算它的步数
steps_for_8888 = collatz_steps(8888)
print(f"Steps for 8888: {steps_for_8888}")

# 计算1到10000的每个数的步数
steps_for_numbers = {n: collatz_steps(n) for n in range(1, 10001)}

# 找到经历步数最多的数
max_steps = max(steps_for_numbers.values())
numbers_with_max_steps = [n for n, steps in steps_for_numbers.items() if steps == max_steps]
print(f"Numbers with the most steps ({max_steps}): {numbers_with_max_steps}")
```

1. 对于给定的数：8888，步数为34。
2. 在1到10000之内，经历步数最多的数是6171，它需要经历261步才能达到1。这个结果再次验证了科拉兹猜想对于已经检验过的范围内的数是成立的。
