# 方程求解

## 方程求解法一之解析方法

解析方法是一种传统的数学方法，用于**精确**求解方程。这种方法依赖于数学公式和定理来直接找出方程的根（解）。对于不同类型的方程，有各自的解析解法。以下是几种常见方程及其解析求解方法：

### 1. 一元一次方程（线性方程）

一元一次方程的标准形式是 $ax + b = 0$，其中 $a$ 和 $b$ 是已知常数，$x$ 是未知数。这种方程的解析解很简单，可以直接通过代数变换求得：

$$x = -\frac{b}{a}$$

### 2. 一元二次方程

一元二次方程的标准形式是 $ax^2 + bx + c = 0$。这种方程可以通过配方法、因式分解或使用著名的二次公式求解。二次公式为：

$$x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$$

### 3. 方程组

方程组的求解方法多种多样，包括代数法（如代入法、消元法）和几何法。对于线性方程组，还可以使用矩阵方法（如高斯消元法或克莱姆法则）进行求解。

### 4. 高阶多项式方程

对于三次方程和四次方程，存在特定的解析求解公式，但这些公式相对复杂。而五次及以上的多项式方程，根据阿贝尔-鲁菲尼定理（Abel-Ruffini theorem），我们知道它们通常没有通用的解析解法。

### 解析方法的局限性

解析方法能提供精确解，但它们的**应用范围有限**，特别是对于非线性方程、高阶多项式方程以及一些特殊类型的方程，可能**不存在解析解**，或者**解析解难以找到**。这时，数学家和工程师会转向数值解法，这种方法通过**迭代计算来逼近**方程的根，虽然可能不是完全精确的，但在工程和科学计算中足够使用。

### 计算机代数系统

对于解析解的求解，计算机代数系统（比如Python）提供了强大的工具（各种库）。这些库能自动应用复杂的数学公式和算法，求解各种类型的方程，包括能找到解析解的方程。它们对于处理复杂的数学运算非常有用，可以大大简化求解过程，让我们专注于问题的分析而不是求解过程的繁琐计算。

```python
# 解一元二次方程组
import sympy as sp 

x = sp.Symbol('x')
a, b, c = 1, -3, 2
equation = a * x ** 2 + b * x + c
result = sp.solve(equation, x)
print(result)
```

```python
# 解方程组

x, y = sp.symbols('x y')
equations = [2 * x - y - 3, 3 * x + y - 7]
result = sp.solve(equations, (x, y))
print(result)
```

## 方程求解法二之图形放大

图形放大法是另一种数值分析方法，用于找出方程的根，即求解 $f(x) = 0$ 的 $x$ 值。这种方法的核心在于使用图形工具绘制函数 $f(x)$ 的图形，通过观察图形与 $x$ 轴的交点来估计方程的根。

### 实现步骤

1. **建立坐标系，画出函数 $f(x)$ 的图形：** 使用计算机图形工具（如 Python 中的 matplotlib 库）来绘制函数曲线。
2. **观察曲线与 $x$ 轴的交点：** 这些交点即为方程的根。
3. **局部放大交点：** 如果需要更精确地确定根的位置，可以在感兴趣的交点附近对图形进行放大，这有助于更清晰地看到交点的具体位置。
4. **估计根的位置：** 根据放大后的图形，估计交点的横坐标值，这个值就是方程的一个根。

### 优点

- **直观性：** 图形放大法直观地展示了函数与 $x$ 轴的交点，使得寻找方程的根变得直观易懂。
- **适用性：** 对于一些解析求解困难的方程，图形方法提供了一种可视化的解决方案。

### 缺点

- **低精确度：** 由于计算机图形的局限性，该方法很难获得根的精确值。计算机上的图形是由**离散点连线**得到的，而非**完美的连续曲线**。
- **低效率：** 对于需要高精度解的问题或大量数据处理，图形放大法可能不是最高效的选择。

### 使用场合

图形放大法特别适用于对方程解的大致位置有一个直观了解的场合，特别是当方程难以解析求解时。它为后续使用更精确的数值方法（如牛顿法、二分法等）提供了一个良好的起点。

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['font.family'] = 'Palatino Linotype'

def f(x):
    return x ** 5 + 2 * x ** 2 + 4

x = np.linspace(-3, 3, 400)
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r'$f(x)=x^5+2x^2+4$', color='r')
plt.xlabel('x', fontsize=20)
plt.ylabel('f(x)', fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.title(r'$f(x) = x^5 + 2x^2 + 4$', fontsize=16)
plt.legend(fontsize=16)
plt.show()
```

## 方程求解法三之迭代法

迭代法是一种寻找方程 $f(x) = 0$ 根的数学方法，它通过重复应用一个迭代过程来逐渐逼近方程的解。这个方法基于将原方程转化为等价形式 $x = j(x)$，其中 $j(x)$ 是一个在给定区间内具有良好性质的函数。这种转化使得原问题转变为寻找函数 $j(x)$ 的不动点，即那些满足 $x^* = j(x^*)$ 的点 $x^*$。

### 基本理论

迭代法的基本理论是基于不动点定理。该定理指出：在某些条件下，函数 $j(x)$ 的连续性和映射特性保证了不动点的存在性和唯一性。迭代法的关键在于**如何选择合适的迭代函数** $j(x)$ 以及**如何确保迭代过程的收敛性**。

### 方法步骤

1. **转化方程：** 将原始方程 $f(x) = 0$ 转化为 $x = j(x)$ 形式。
2. **选择初值：** 根据问题的具体情况，选择一个合理的初始迭代值 $x_0$。
3. **迭代过程：** 应用迭代公式 $x_{n+1} = j(x_n)$ 进行迭代。
4. **检查收敛：** 判断迭代是否收敛，即是否满足 $|x_{n+1} - x_n| < \varepsilon$，其中 $\varepsilon$ 是预先设定的一个小的正数，代表容许误差。
5. **终止条件：** 当达到收敛条件或迭代次数超过预定值时，停止迭代。

### 注意事项

- **收敛性：** 并非所有的迭代函数 $j(x)$ 都能保证迭代过程的收敛。**选择一个合适的迭代函数和初值**是迭代法成功的关键。
- **速度：** 不同的迭代函数或初值会影响迭代的速度。理想的迭代函数应该能**快速**收敛到方程的根。

### 应用场景

迭代法特别适用于那些难以直接求解或无法通过解析方法求解的方程。通过适当选择迭代函数和初值，迭代法可以有效地逼近方程的根。

### 举个栗子

假设我们有方程 $x^2 - x - 2 = 0$，我们可以将其转化为迭代形式 $x = \sqrt{x + 2}$ 或 $x = 1 + \frac{2}{x}$ 等形式。选择一个合适的初值，如 $x_0 = 1$，然后应用迭代公式进行迭代，直到满足收敛条件。

迭代法的实现需要细心选择迭代函数和初值，并且要有适当的收敛性检验。正确应用迭代法可以高效地求解许多数学和工程问题中的方程。

### 牛顿法（Newton's method）

牛顿法（Newton's method），也称为牛顿-拉弗森方法（Newton-Raphson method），是一种在实数和复数上寻找方程根的迭代法。牛顿法使用函数 $f(x)$及其导数 $f'(x)$ 来快速逼近方程的根。其基本迭代公式为：

$$x_{k+1} = x_k - \frac{f(x_k)}{f'(x_k)}$$

牛顿法的优点在于收敛速度快（局部二次收敛）和通用性好，但它要求函数在求解的区域内可微，并且能够计算出函数的导数。

下面将用 Python 的 `SymPy` 库来实现牛顿法，求解方程 $x^3 - x^2 - x - 1 = 0$ 的根。

```python
import sympy as sp 

x = sp.symbols('x')

f = x ** 3 - x ** 2 - x - 1
f_prime = sp.diff(f, x)

# 定义牛顿法的迭代函数
def newton_method(func, func_prime, initial_guess, tolerance=1e-6, max_iterations=100):
    xk = initial_guess
    for _ in range(max_iterations):
        xk_new = xk - func.subs(x, xk) / func_prime.subs(x, xk)
        if abs(xk_new - xk) < tolerance:
            return xk_new
        xk = xk_new
    return xk

initial_guess = 2.0
root = newton_method(f, f_prime, initial_guess)

print(f"The root of the equation is approximately: {root:.3f}")
```

### 二分法（Bisection method）

二分法（Bisection method）是一种基于中间值定理的根查找算法，用于求解在闭区间 $[a, b]$ 上连续函数 $f(x)$ 的零点。二分法的核心思想是，如果 $f(a)$ 和 $f(b)$ 异号（即一个正一个负，表明 $f(x)$ 在 $a$ 和 $b$ 之间至少有一个根），则根据中间值定理，函数在这个区间内至少有一个零点。

#### 基本步骤

1. **选择初始区间 $[a, b]$，确保 $f(a) \cdot f(b) < 0$**，即端点处的函数值异号。
2. **计算中点 $c = \frac{a + b}{2}$ 的函数值 $f(c)$。**
3. **判断 $f(c)$ 的符号：**
   - 如果 $f(c) = 0$，则 $c$ 就是一个零点。
   - 如果 $f(c) \cdot f(a) < 0$，则零点在 $[a, c]$ 内，更新 $b = c$。
   - 如果 $f(c) \cdot f(b) < 0$，则零点在 $[c, b]$ 内，更新 $a = c$。
4. **重复步骤 2 和 3，直到满足停止条件，通常是 $|b - a| < \varepsilon$，其中 $\varepsilon$ 是预设的精度要求。**

#### 优点

- **简单易懂：** 二分法的原理简单，易于实现。
- **稳定可靠：** 只要初始区间内确实存在零点，二分法就能保证找到解。
- **不需导数：** 与牛顿法等需要函数导数的方法不同，二分法不需要函数的导数信息，适用范围更广。

#### 缺点

- **收敛速度：** 相对于牛顿法和其他迭代法，二分法的收敛速度较慢。
- **初始区间的选择：** 需要事先知道一个包含零点的区间，对于一些函数可能难以确定。

### 1. 求下列方程的根

#### 1.1 求方程 $x \sin(x) + \cos(x) - \sin(x) - 2x = 0$ 在 $[-1, 1]$ 上的近似解

可以使用二分法来解决这个问题，因为二分法适用于在**指定区间内**查找**连续函数**的根。首先定义函数 $f(x) = x \sin(x) + \cos(x) - \sin(x) - 2x$，然后应用二分法。

通过代码求得，方程 $x \sin(x) + \cos(x) - \sin(x) - 2x = 0$ 在 $[-1, 1]$ 上的根大约是 $0.356$。

```python
import numpy as np

def f(x):
    return x * np.sin(x) + np.cos(x) - np.sin(x) - 2 * x

def bisection_method(f, a, b, tol=1e-6):
    if f(a) * f(b) > 0:
        raise ValueError("The function values at the end points must have opposite signs")
    while (b - a) / 2 > tol:
        midpoint = (a + b) / 2
        if f(midpoint) == 0:
            return midpoint
        elif f(a) * f(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint
    return (a + b) / 2

a = -1
b = 1
root = bisection_method(f, a, b)
print(f"The root of the equation is approximately: {root:.3f}")
```

#### 1.3：找出函数 $f(x) = x^3 - 6x^2 - 2x + 12$ 的所有零点

由于这个函数是一个多项式函数，我们可以使用 `numpy` 的 `roots` 函数来找到它的所有零点。这个方法不依赖于图形化分析，而是直接计算多项式的根。

通过代码实现，可以看出函数 $f(x) = x^3 - 6x^2 - 2x + 12$ 的所有零点是 $x = 6$、$x = -1.414$ 和 $x = 1.414$。

### 2. 求解线性方程组

$$
\begin{align*}
7x_1 + 2x_2 + x_3 - 2x_4 &= 4 \\
9x_1 + 15x_2 + 3x_3 - 2x_4 &= 7 \\
-2x_1 - 2x_2 + 11x_3 + 5x_4 &= -1 \\
x_1 + 3x_2 + 2x_3 + 13x_4 &= 0
\end{align*}
$$

```python
A = np.array([[7, 2, 1, -2],
              [9, 15, 3, -2], 
              [-2, -2, 11, 5],
              [1, 3, 2, 13]])

b = np.array([4, 7, -1, 0])

solution = np.linalg.solve(A, b)
print(f"The solution to the equation is: {solution}")
```

### 3. 求下列方程组的解

这两个方程组均属于非线性方程组，可以使用**数值方法**进行求解。对于这类问题，常用的方法包括牛顿法（或牛顿-拉弗森方法）和使用 `scipy.optimize` 库中的 `fsolve` 函数。

#### 3.1 求解（3.1）方程组

$$
\begin{equation*}
\begin{cases}
2x_1 - x_2 = e^{-x_1} \\
-x_1 + 2x_2 = e^{-x_2}
\end{cases}
\tag{3.1}
\end{equation*}
$$

```python
from scipy.optimize import fsolve

def equations(p):
    x1, x2 = p
    return (2 * x1 - x2 - np.exp(-x1), -x1 + 2 * x2 - np.exp(-x2))

initial_guess = [-5, -5]
result = fsolve(equations, initial_guess)
print(f"The solution to the equation is: x1 = {result[0]:.3f}, x2 = {result[1]:.3f}")
```

#### 3.2 求解（3.2）方程组

$$
\begin{equation*}
\begin{cases}
x_1^2 - 5x_2^2 + 7x_3^2 = -12 \\
3x_1 x_2 + x_1 x_3 - 11x_1 = 0 \\
2x_2 x_3 + 40x_1 = 0
\end{cases}
\tag{3.2}
\end{equation*}
$$

```python
def equations(p):
    x1, x2, x3 = p
    return (x1 ** 2 - 5 * x2 ** 2 + 7 * x3 ** 2 + 12,
            3 * x1 * x2 + x1 * x3 - 11 * x1,
            2 * x2 * x3 + 40 * x1)

initial_guess = [1, 1, 1]
solution = fsolve(equations, initial_guess)
print(f"The solution to the equation is: x1 = {solution[0]:.3f}, x2 = {solution[1]:.3f}, x3 = {solution[2]:.3f}")
```
