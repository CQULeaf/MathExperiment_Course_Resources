## 一些简单的函数曲线的绘制

图中展示了四个不同的数学函数，分别是：

$$ y_1 = x $$
$$ y_2 = x - \frac{x^3}{3!} $$
$$ y_3 = x - \frac{x^3}{3!} + \frac{x^5}{5!} $$
$$ y_4 = \sin(x) $$

这些函数代表了Taylor级数在不同阶数的近似。其中 $ y_1 $ 是最简单的线性函数，$ y_2 $ 和 $ y_3 $ 是函数 $ \sin(x) $ 的二阶和四阶Taylor级数近似。而 $ y_4 $ 是实际的 $ \sin(x) $ 函数。随着级数项的增加，Taylor级数近似会越来越接近实际的 $ \sin(x) $ 函数。

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['font.family'] = 'Palatino Linotype'

x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

y1 = x
y2 = x - (x**3) / math.factorial(3)
y3 = x - (x**3) / math.factorial(3) + (x**5) / math.factorial(5)
y4 = np.sin(x)

plt.figure(figsize = (10, 6))
plt.plot(x, y1, label='y1 = x', color = '#EAAA60', linewidth=2)
plt.plot(x, y2, label='y2 = x - x^3/3!', color = '#B7B2D0', linewidth=2)
plt.plot(x, y3, label='y3 = x - x^3/3! + x^5/5!', color = '#84C3B7', linewidth=2)
plt.plot(x, y4, label='y4 = sin(x)', color = '#E68B81', linewidth=2)

plt.title('Comparison of Series Approximations with sin(x)', fontsize = 16)
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)
plt.legend(fontsize = 15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True)
plt.show()
```

## 一些复杂的二维曲线的绘制

四个子图分别呈现了四种不同的数学曲线的可视化图形，每个子图展示的曲线及其对应的数学公式如下：

1. 概率密度曲线：
$$ y = e^{-x^2} $$
这是正态分布的概率密度函数的一个部分，特别是标准正态分布（均值为0，方差为1）的概率密度函数。

2. 四叶玫瑰线（极坐标）：
$$ r = \sin(2\theta) $$
这是极坐标下的四叶玫瑰线方程，当θ从0变化到2π时，曲线在极坐标平面上绘制出四个对称的“花瓣”。

3. 双扭线（隐函数）：
$$ x^3 + y^3 - 3axy = 0 $$
这是笛卡尔双扭线的隐函数方程。在代码中，参数 `a` 被隐含地设置为1。双扭线是一种代数曲线，通常以其对称性和独特的形状而知名。

4. 曳物线：
$$ y = \cosh(x) - 1 $$
这是曳物线的方程，其中 `cosh` 是双曲余弦函数。曳物线是悬链线的形状，也就是一个理想的柔性不可伸缩链条两端固定时的形状。

```python
plt.figure(figsize=(14, 8))

mpl.rcParams['font.family'] = 'Palatino Linotype'

# 1. 概率密度曲线
plt.subplot(2, 2, 1)
x = np.linspace(-3, 3, 1000)
y = np.exp(-x**2)  # 正态分布的概率密度公式
plt.plot(x, y, color = '#A2AFC4')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title('Probability Density Curve', fontsize=16)

# 2. 四叶玫瑰线 (极坐标)
plt.subplot(2, 2, 2, projection='polar')
theta = np.linspace(0, 2*np.pi, 1000)
r = np.sin(2 * theta) # 极坐标下的四叶玫瑰线方程
plt.plot(theta, r, color = '#78837E')
plt.tick_params(labelsize=15)
plt.title('Four-Leaf Rose (Polar Coordinates)', fontsize=16)

# 3. 双扭线 (隐函数)
plt.subplot(2, 2, 3)
# 使用隐函数绘制双扭线的标准形式 x^3 + y^3 - 3axy = 0
x = np.linspace(-2, 2, 1000)
y = np.linspace(-2, 2, 1000)
X, Y = np.meshgrid(x, y)
Z = X**3 + Y**3 - 3*X*Y  # 双扭线的隐函数方程
# 绘制等高线，其中Z=0的等高线即为双扭线
plt.contour(X, Y, Z, levels=[0])
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title('Folium of Descartes (Standard Form)', fontsize=16)

# 4. 曳物线
plt.subplot(2, 2, 4)
x = np.linspace(-2, 2, 1000)
y = np.cosh(x) - 1 # 曳物线方程
plt.plot(x, y, color = '#7DA6C6')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title('Catenary Curve', fontsize=16)

plt.tight_layout()
plt.show()
```

## 三维曲面图形的可视化

图中可视化了两组方程。左图方程是一个三维空间曲面的参数方程。右图方程是一个单独的方程，定义了z轴的值作为x和y的函数。

1. 左图参数方程定义了一组变量x，y和z作为u和v的函数，其中u和v都在$ (0,2\pi) $的范围内。这些方程如下：
$$
\begin{cases}
x = (1 + \cos u) \cos v \\
y = (1 + \cos u) \sin v \\
z = \sin u
\end{cases}
$$

2. 右图方程是：
$$ z = \sin(\pi \sqrt{x^2 + 4y^2}) $$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

mpl.rcParams['font.family'] = 'Palatino Linotype'

u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, 2 * np.pi, 100)
u, v = np.meshgrid(u, v)

x = (1 + np.cos(u)) * np.cos(v)
y = (1 + np.cos(u)) * np.sin(v)
z = np.sin(u)

fig = plt.figure(figsize=(16, 8))

# 绘制第一幅子图
ax1 = fig.add_subplot(121, projection='3d')
ax1.set_title("Parametric Surface", fontsize=20)
ax1.plot_surface(x, y, z, rstride=5, cstride=5, color='skyblue', edgecolors='w', alpha=0.7)
ax1.set_xlabel('X axis', fontsize=18)
ax1.set_ylabel('Y axis', fontsize=18)
ax1.set_zlabel('Z axis', fontsize=18)

# 绘制第二幅子图
ax2 = fig.add_subplot(122, projection='3d')
ax2.set_title("z = sin(pi * sqrt(x^2 + 4y^2))", fontsize=20)

# Create a grid for x and y
x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)
x, y = np.meshgrid(x, y)
z = np.sin(np.pi * np.sqrt(x**2 + 4*y**2))

# Plot the surface
ax2.plot_surface(x, y, z, rstride=20, cstride=20, color='lightgreen', edgecolors='w', alpha=0.7)
ax2.set_xlabel('X axis', fontsize=18)
ax2.set_ylabel('Y axis', fontsize=18)
ax2.set_zlabel('Z axis', fontsize=18)

# Show the plot
plt.tight_layout()
plt.show()
```
